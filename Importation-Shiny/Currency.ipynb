```python
import numpy as np
import pandas as pd

data = pd.read_csv('Final_Dataset_Location.csv')
data.drop('Unnamed: 0', axis = 1, inplace = True)
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Customs_Value_PHP'] = data['Customs_Value'] * data['Exchange_Rate']
```

```python
exchange = data[['Currency', 'Exchange_Rate', 'Date']]
exchange = exchange[exchange['Currency'] == 'USD']
exchange.reset_index(drop = True, inplace = True)
exchange = exchange[['Exchange_Rate', 'Date']].groupby('Date').mean()
```

```python
datetime_index = pd.date_range(start='2016-01-01', end='2019-07-01', freq = 'D')[:-1]
merge_byday = pd.DataFrame()
merge_byday['Date'] = datetime_index
merge_byday = merge_byday.merge(exchange, how = 'left', left_on = 'Date', right_on = 'Date')
```

```python
merge_byday = merge_byday.ffill(axis = 0) 
merge_byday.iloc[0,1] = 47.166
merge_byday
```

```python
merge_byday.columns = ['Date', 'USD_to_PHP']
```

```python
data = data.merge(merge_byday, how = 'left', left_on = 'Date', right_on = 'Date')
```
